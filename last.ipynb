```python
import os
import subprocess

VIDEO_BASE = "/home/Pdanova/Dataset/Dataset/AVEC2014_AudioVisual/Video/Development"
AUDIO_BASE = "/home/Pdanova/Dataset/Dataset/AVEC2014_AudioVisual/Audio/Development"

tasks = ["Freeform", "Northwind"]

os.makedirs(AUDIO_BASE, exist_ok=True)

for task in tasks:
    video_dir = os.path.join(VIDEO_BASE, task)
    audio_dir = os.path.join(AUDIO_BASE, task)
    os.makedirs(audio_dir, exist_ok=True)

    print(f"\n=== Processing {task} ===")

    for fname in os.listdir(video_dir):
        if fname.endswith("_video.mp4"):
            prefix = fname.replace("_video.mp4", "")

            mp4_path = os.path.join(video_dir, fname)
            wav_path = os.path.join(audio_dir, prefix + "_audio.wav")

            # ì´ë¯¸ ì¡´ìž¬í•˜ë©´ skip
            if os.path.exists(wav_path):
                print(f"[SKIP] {wav_path} already exists")
                continue

            # ffmpeg ë³€í™˜
            cmd = [
                "ffmpeg", "-y",
                "-i", mp4_path,
                "-vn",
                "-ac", "1",
                "-ar", "16000",
                "-f", "wav",
                wav_path
            ]

            print(f"[CONVERT] {mp4_path} â†’ {wav_path}")
            subprocess.run(cmd, stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)

print("\nðŸŽ‰ All conversions completed.")
```

```python
import os
import cv2
import librosa
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader
import timm
import pandas as pd
from tqdm import tqdm
from scipy.stats import pearsonr

device = "cuda" if torch.cuda.is_available() else "cpu"


###############################################################
# 1. Load N frames from video (7 frames)
###############################################################
def load_video_frames(path, num_frames=7):
    cap = cv2.VideoCapture(path)
    total = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    idx = np.linspace(0, max(total - 1, 0), num_frames, dtype=int)

    frames = []
    for i in idx:
        cap.set(cv2.CAP_PROP_POS_FRAMES, i)
        ret, f = cap.read()
        if not ret:
            print(f"[WARN] dummy frame used! path={path}, frame_idx={i}, total={total}")
            f = np.zeros((299, 299, 3), np.uint8)  # ì‹¤íŒ¨ ì‹œ ë”ë¯¸ í”„ë ˆìž„
        f = cv2.cvtColor(f, cv2.COLOR_BGR2RGB)
        f = cv2.resize(f, (299, 299))
        frames.append(f)

    cap.release()
    frames = np.stack(frames)
    frames = torch.tensor(frames).permute(0, 3, 1, 2).float() / 255.
    return frames



###############################################################
###############################################################
# 2. Load MFCC (Audio) - ê³ ì • ê¸¸ì´ë¡œ íŒ¨ë”©/ìžë¥´ê¸° ì¶”ê°€
###############################################################
def load_mfcc(path, n_mfcc=40, target_len=3000):
    y, sr = librosa.load(path, sr=16000)  #ìƒ˜í”Œë§ ë ˆì´íŠ¸ë¥¼ 16kHz

    mel = librosa.feature.melspectrogram(
        y=y, sr=sr, n_mels=128, fmax=8000  #ìµœëŒ€ ì£¼íŒŒìˆ˜ 8kHz
    )
    mel_db = librosa.power_to_db(mel, ref=np.max)

    mfcc = librosa.feature.mfcc(
        S=mel_db,
        n_mfcc=n_mfcc
    )  # (40, T)

    # --- ê¸¸ì´ ë§žì¶”ê¸° ---
    T = mfcc.shape[1]

    if T < target_len:
        pad_width = target_len - T
        mfcc = np.pad(mfcc, ((0, 0), (0, pad_width)), mode="constant")
    else:
        mfcc = mfcc[:, :target_len]

    return torch.tensor(mfcc, dtype=torch.float32)


###############################################################
# 3. Dataset
###############################################################
class AVECDataset(Dataset):
    def __init__(self, video_root, audio_root, label_root):
        ff_dir = os.path.join(video_root, "Freeform")
        self.prefixes = sorted([f.split("_Freeform")[0] for f in os.listdir(ff_dir)])

        self.samples = []
        for p in self.prefixes:
            v = os.path.join(video_root, "Freeform", f"{p}_Freeform_video.mp4")
            a = os.path.join(audio_root, "Freeform", f"{p}_Freeform_audio.wav")
            l = os.path.join(label_root, f"{p}_Depression.csv")

            if os.path.exists(v) and os.path.exists(a) and os.path.exists(l):
                self.samples.append((v, a, l))

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        vpath, apath, lpath = self.samples[idx]
        frames = load_video_frames(vpath, num_frames=7)  
        mfcc = load_mfcc(apath)

        df = pd.read_csv(lpath, header=None)
        label = torch.tensor(float(df.iloc[0, 0])).float()
        return frames, mfcc, label


###############################################################
# 4. Kirsch Edge Detection
###############################################################
KIRSCH_KERNELS = torch.tensor([
    [[5, 5, 5], [-3, 0, -3], [-3, -3, -3]],
    [[5, 5, -3], [5, 0, -3], [-3, -3, -3]],
    [[5, -3, -3], [5, 0, -3], [5, -3, -3]],
    [[-3, -3, -3], [5, 0, -3], [5, 5, -3]],
    [[-3, -3, -3], [-3, 0, -3], [5, 5, 5]],
    [[-3, -3, -3], [-3, 0, 5], [-3, 5, 5]],
    [[-3, -3, 5], [-3, 0, 5], [-3, -3, 5]],
    [[-3, 5, 5], [-3, 0, 5], [-3, -3, -3]]
], dtype=torch.float32).unsqueeze(1)

def kirsch_edge(x):
    gray = x.mean(1, keepdim=True)
    responses = [F.conv2d(gray, k.unsqueeze(0).to(x.device), padding=1)
                 for k in KIRSCH_KERNELS]
    edge = torch.max(torch.stack(responses), dim=0).values
    return edge.mean((2, 3))  # (B, 1)


###############################################################
# 5. Temporal Difference
###############################################################
def temporal_diff(f0, f1):
    return torch.abs(f1 - f0).mean((2, 3))  # (B, 3)


###############################################################
# 6. Audio Encoder (Residual + LSTM + FCÃ—2)
###############################################################
class ResidualBlock1D(nn.Module):
    def __init__(self, in_ch, out_ch):
        super().__init__()
        self.conv1 = nn.Conv1d(in_ch, out_ch, 3, padding=1)
        self.conv2 = nn.Conv1d(out_ch, out_ch, 3, padding=1)
        self.short = nn.Conv1d(in_ch, out_ch, 1)

    def forward(self, x):
        out = F.relu(self.conv1(x))
        out = self.conv2(out)
        return F.relu(out + self.short(x))


class AudioEncoder(nn.Module):
    def __init__(self):
        super().__init__()
        self.res1 = ResidualBlock1D(40, 128)
        self.res2 = ResidualBlock1D(128, 256)

        self.lstm = nn.LSTM(
            input_size=256,
            hidden_size=256,
            batch_first=True,
            bidirectional=True
        )

        self.fc1 = nn.Linear(512, 256)
        self.fc2 = nn.Linear(256, 256)

    def forward(self, mfcc):
        if mfcc.dim() == 2:
            mfcc = mfcc.unsqueeze(0)

        x = self.res1(mfcc)
        x = self.res2(x)
        x = x.permute(0, 2, 1)

        x, _ = self.lstm(x)
        x = x[:, -1]

        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x  # (B, 256)


###############################################################
# 7. Video Encoder (Inception-ResNet + Generic T-frame Seq)
###############################################################
class VideoEncoder(nn.Module):
    def __init__(self):
        super().__init__()
        self.backbone = timm.create_model(
            "inception_resnet_v2.tf_in1k",
            pretrained=True
        )
        self.backbone.classifier = nn.Identity()

        self.pool = nn.AdaptiveAvgPool2d((1, 1))
        self.fc = nn.Linear(1536, 256)  # frame feature â†’ 256D

    def encode_frame(self, f):
        # f: (B*T, 3, 299, 299)
        x = self.backbone.forward_features(f)
        x = self.pool(x).view(x.size(0), -1)
        return self.fc(x)  # (B*T, 256)

    def forward(self, frames):
        """
        frames: (B, T, 3, 299, 299) 
        TëŠ” 4ë“  8ì´ë“  ìƒê´€ì—†ì´ ì²˜ë¦¬ ê°€ëŠ¥
        """
        B, T, C, H, W = frames.shape

        # 1) ëª¨ë“  í”„ë ˆìž„ í•œ ë²ˆì— backbone í†µê³¼ (íš¨ìœ¨â†‘)
        flat_frames = frames.view(B * T, C, H, W)              # (B*T, 3, 299, 299)
        A_flat = self.encode_frame(flat_frames)                # (B*T, 256)
        A = A_flat.view(B, T, 256)                             # (B, T, 256)

        # 2) Kirsch edge (ê° í”„ë ˆìž„ë³„ edge ê°•ë„ 1D)
        edges_flat = kirsch_edge(flat_frames)                  # (B*T, 1)
        E = edges_flat.view(B, T, 1)                           # (B, T, 1)

        # 3) Temporal diff: ì¸ì ‘ í”„ë ˆìž„ ê°„ ì°¨ì´ (B, T, 3)
        diffs = []
        for t in range(T):
            if t == T - 1:
                # ë§ˆì§€ë§‰ í”„ë ˆìž„ì˜ diffëŠ” ì§ì „ diffë¥¼ ìž¬ì‚¬ìš©í•˜ê±°ë‚˜ 0ìœ¼ë¡œ ë‘ 
                if T > 1:
                    diffs.append(diffs[-1])  # ì§ì „ diff ë³µì‚¬
                else:
                    diffs.append(torch.zeros(B, 3, device=frames.device))
            else:
                d = temporal_diff(frames[:, t], frames[:, t + 1])  # (B, 3)
                diffs.append(d)
        D = torch.stack(diffs, dim=1)                             # (B, T, 3)

        # 4) [Appearance(256) + Edge(1) + Motion(3)] = 260D ì‹œí€€ìŠ¤
        seq = torch.cat([A, E, D], dim=2)                         # (B, T, 260)
        return seq

        
###############################################################
# 8. TAP (Temporal Attentive Pooling)
###############################################################
class TAP(nn.Module):
    def __init__(self, dim):
        super().__init__()
        self.U = nn.Linear(dim, dim)
        self.W = nn.Linear(dim, 1)

    def forward(self, H):  # (B, T, D)
        score = self.W(torch.tanh(self.U(H))).squeeze(-1)  # (B, T)
        attn = torch.softmax(score, dim=1).unsqueeze(-1)   # (B, T, 1)
        Z = (attn * H).sum(dim=1)                          # (B, D)
        return Z


###############################################################
# 9. Multi-modal Fusion Model
###############################################################
class MultiModalNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.video = VideoEncoder()
        self.audio = AudioEncoder()

        self.tap = TAP(260 + 256)  # dim = 516
        self.reg = nn.Linear(260 + 256, 1)

    def forward(self, frames, mfcc):
        V = self.video(frames)     # (B, T, 260)  â† ì´ì œ T=8
        A = self.audio(mfcc)       # (B, 256)

        A_expand = A.unsqueeze(1).repeat(1, V.size(1), 1)  # (B, T, 256)
        H = torch.cat([V, A_expand], dim=2)                # (B, T, 516)

        fused = self.tap(H)                                # (B, 516)
        return self.reg(fused)


###############################################################
# 10. Pearson Loss
###############################################################
def pearson_loss(pred, target):
    vx = pred - pred.mean()
    vy = target - target.mean()
    denom = torch.sqrt((vx**2).sum()) * torch.sqrt((vy**2).sum())
    if denom == 0:
        return 0
    return 1 - (vx * vy).sum() / denom


###############################################################
# 11. Training Loop
###############################################################
video_root = "/home/Pdanova/Dataset/Dataset/AVEC2014_AudioVisual/Video/Training"
audio_root = "/home/Pdanova/Dataset/Dataset/AVEC2014_AudioVisual/Audio/Training"
label_root = "/home/Pdanova/Dataset/Dataset/Final_Labels/Training_DepressionLabels"


train_set = AVECDataset(video_root, audio_root, label_root)

train_loader = DataLoader(
    train_set,
    batch_size=2,
    shuffle=True,
    num_workers=0,   
    pin_memory=True
)


model = MultiModalNet().to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=5e-5)
scaler = torch.cuda.amp.GradScaler()

EPOCHS = 18
for epoch in range(EPOCHS):

    model.train()
    preds_epoch, gts_epoch = [], []
    total_loss = 0

    loop = tqdm(train_loader, desc=f"Epoch {epoch+1}/{EPOCHS}")

    for frames, mfcc, labels in loop:
        frames, mfcc, labels = (
            frames.to(device),
            mfcc.to(device),
            labels.to(device)
        )

        optimizer.zero_grad()
        with torch.cuda.amp.autocast():
            out = model(frames, mfcc).squeeze()
            loss = F.mse_loss(out, labels) + 0.2 * pearson_loss(out, labels)

        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()

        total_loss += loss.item()
        preds_epoch.extend(out.detach().cpu().numpy())
        gts_epoch.extend(labels.detach().cpu().numpy())

        loop.set_postfix(loss=loss.item())

    preds = np.array(preds_epoch)
    gts = np.array(gts_epoch)

    mae = np.mean(np.abs(preds - gts))
    rmse = np.sqrt(np.mean((preds - gts) ** 2))
    corr = pearsonr(preds, gts)[0]

    print(f"\n[Epoch {epoch+1}] "
          f"Loss={total_loss/len(train_loader):.4f} | MAE={mae:.3f} | RMSE={rmse:.3f} | Corr={corr:.3f}")
```
